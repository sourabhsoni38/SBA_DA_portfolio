### Cleaning Dataset

```python
df['LookupCodes'] = df['NAICS Industry Description'].str.slice(start = 7, stop = 9)
```

```python
df.head()
```

```python
df['NAICS Industry Description'].str.len()
```

```python
df['Sector'] = df['NAICS Industry Description'].str.slice(start = 12, stop = 550)
```

```python
df.head()
```

```python
df = df[['NAICS Industry Description', 'LookupCodes', 'Sector']]
```

```python
df.info()
```

```python
df[df['LookupCodes'].str.isnumeric()]
```

```python
df.loc[df['Sector'] == '33 – Manufacturing', 'Sector'] = 'Manufacturing'
```

```python
df
```

```python
df['Sector'] = np.where(df['Sector'] == '45 – Retail Trade', 'Retail Trade', df['Sector'])
df['Sector'] = np.where(df['Sector'] == '49 – Transportation and Warehousing', 'Transportation and Warehousing', df['Sector'])
```

```python
df
```

```python
df.loc [len(df.index)] = ['Sector 31 – 33 – Manufacturing', 32, 'Manufacturing']
df.loc [len(df.index)] = ['Sector 31 – 33 – Manufacturing', 33, 'Manufacturing']

df2 = {'NAICS Industry Description': 'Sector 44 - 45 – Retail Trade', 'LookupCodes': 45, 'Sector': 'Retail Trade'}
df = df.append(df2, ignore_index = True)
```

```python
df.loc[len(df.index)] = ['Sector 48 - 49 – Transportation and Warehousing', 49, 'Transportation and Warehousing']
```

```python
# export cleaned datset into another csv file

df.to_csv('Clean_data_without_index.csv', encoding = 'utf-8', index = False)
```
